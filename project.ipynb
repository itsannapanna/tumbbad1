DEFINING THE COST FUNCTION 

```python
def calc_cost(x, y, w, b): 
    
    # number of training examples
    m = x.shape[0] 
    
    cost = 0
    for i in range(m):
        y_pred = np.dot(w,x[i]) + b 
        cost += np.sum((y_pred - y[i]) ** 2)
        
    total_cost = (cost / (2*m) ) 
    return total_cost
```

```python
FUNCTION FOR GRADIENT DESCENT
```

```python
def compute_gradient(x, y, w, b): 
    
    # Number of training examples
    m = x.shape[0]
    
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = np.dot(w,x[i]) + b 
        dj_dw_i = (f_wb - y[i])*x[i]
        dj_db_i = (f_wb - y[i])
        dj_dw += (dj_dw_i) 
        dj_db += (dj_db_i) 
        
    dj_dw = dj_dw / m
    dj_db = dj_db / m
     
    return (dj_dw), (dj_db)
```
